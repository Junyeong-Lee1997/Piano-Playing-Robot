2. 학습된 딥러닝 모델 불러오기

```python
from keras.models import load_model
model = load_model('Diatonic_E150_B64.h5')
model.summary()
```

3. 딥러닝 모델에 넣고 예측되는 코드값 리스트업 하기

```python
#필요한 모듈 임포트

import matplotlib.pyplot as plt
import os
import numpy as np


path2='D:/ai/test_data_set' #테스트 셋 경로 지정
file_list2 = os.listdir(path2) #테스트 셋 경로의 파일들 리스트업
m=len(file_list2)
print('Total Chord Number is %d'%(len(os.listdir('D:/ai/test_data_set'))))

chord=[] #여기에 코드에 대한 정보들을 담을 예정

#모델에 돌려 테스트한 코드에 대한 정보를 출력하여 Chord List에 저장

for i in range(m):
    test_num = plt.imread('D:/ai/test_data_set/ChordImage%d.jpg'%(i+1))
    print('ChordImage%d is '%(i+1), np.argmax(model.predict((test_num.reshape((1, 28, 28, 1)))), axis=-1))
    chord.append(np.argmax(model.predict((test_num.reshape((1, 28, 28, 1)))), axis=-1))
```

4. 코드에 해당하는 UDP 신호 만들기

```python
#필요한 모듈 임포트

import socket
import time

#앞에서 만든 코드 리스트에서 각각을 출력하여 UDP로 3초마다 보냄

for i in range(len(chord)):
    
    sock=socket.socket(socket.AF_INET, socket.SOCK_DGRAM)
    
    #C Diatonic Scale
    
    if chord[i] == 0: #C
        udp='W_17W_26W_28' #코드에 대한 정보 W는 흰 건반, B는 검은 건반 숫자는 건반의 순서(17번째 건반 이런식)
        Ch='C'
        time.sleep(3)
    elif chord[i] == 1: #Dm
        udp='W_18W_27W_29'
        Ch='Dm'
        time.sleep(3)
    elif chord[i] == 2: #Em
        udp='W_19W_28W_30'
        Ch='Em'
        time.sleep(3)
    elif chord[i] == 3: #F
        udp='W_20W_27W_29'
        Ch='F'
        time.sleep(3)
    elif chord[i] == 4: #G7
        udp='W_21W_30W_32'
        Ch='G7'
        time.sleep(3)
    elif chord[i] == 5: #Am
        udp='W_22W_31W_33'
        Ch='Am'
        time.sleep(3)
    elif chord[i] == 6: #Bdim
        udp='W_23W_32W_34'
        Ch='Bdim'
        time.sleep(3)
    else : #Blank
        time.sleep(3)
        Ch='X'
        
    print(Ch)
    sock.sendto(udp.encode(),('192.168.0.6',61557)) #UDP를 받을 컴퓨터의 아이피 주소, 포트 숫자
```
